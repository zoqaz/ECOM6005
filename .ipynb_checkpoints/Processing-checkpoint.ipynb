```python
### Zohaib Qazi
### 19712200
### zohaib@curtin.edu.au

import glob
import pandas as pd
import re
from collections import Counter
import numpy as np
from gensim.models import Word2Vec
import logging
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from gensim.utils import simple_preprocess
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
warnings.filterwarnings("ignore", category=DeprecationWarning) 
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns
import matplotlib
import itertools
matplotlib.style.use('seaborn')
from dateutil import parser
from datetime import date, timedelta
import os

def wordcountfunc(poplist):
    popval = poplist.pop()
    labels, values = zip(*popval.items())
    indSort = np.argsort(values)[::-1]
    labels = np.array(labels)[indSort]
    values = np.array(values)[indSort]
    indexes = np.arange(len(labels))
    return values, labels


class Article:
    def __init__(self, number, ticker, date, title, contents):
        self.number = number
        self.ticker = ticker
        self.date = date
        self.title = title
        self.contents = contents
        self.newswire = False
        self.nwtext = []
        
    def getText(self):
        return self.contents
    
    def makeNewsWire(self):
        self.newswire = True
        
    def getNewsWire(self):
        return self.newswire
    
    def getTicker(self):
        return self.ticker
    
    def getDate(self):
        return self.date
    
    def getNumber(self):
        return self.number
    
    def getTitle(self):
        return self.title
    
    def makeNWText(self, text):
        self.nwtext.append(text)
        
    def getNWText(self):
        return self.nwtext
    
print('done')
```

```python
blah = 'STO'
paths = glob.glob('Articles/'+blah+'/*.txt')
print('done')
```

```python
arts = []

for i in paths:
    i = i.replace("\\","/")
    with open(i, 'r', encoding="utf8") as f:
        lines = f.read()
        split = lines.split(';')
        art = Article(split[0], split[1], split[2], split[3], split[4])
        arts.append(art)
print('COMPLETE')
```

```python
df = pd.DataFrame(data=None, columns = ["Number", "Ticker", "Date", "Year", "Month", "Title", "Wordcount", "Ticker Count",
                                        "3Letter", "4Letter", "Newswire", "Object"])


for i in arts:
    ticker = i.getTicker()
    text = i.getText()
    try:
        os.mkdir('Clusters/'+ticker+"/")
    except FileExistsError:
        pass
    savepath = "Clusters/"+ticker+"/"
    
    x = re.findall(r'(?!AUD|EUR|CHF|USD|SEK|SGD|GBP|JPY|CNY|DKK|AND|THE|FOR|HAS|ARE|ASX|NSW|QLD|VIC|ACT|GMT)\
                    b[A-Z]{1}[A-Z0-9]{2}\b', text)
    y = re.findall(r'\b[A-Z]{1}[A-Z0-9]{3}\b', text)
    tickerRegex = r"\b" + ticker + r"\b"
    z = re.findall(tickerRegex, text)
    
    county = Counter(y).most_common()
    countx = Counter(x).most_common()
    for j in countx:
        if j[0] == ticker:
            countx.remove(j)

    text2 = re.sub(r'[^a-zA-Z]', ' ', text)
    wordcount = text2.split()
    [wordcount.remove(ii) for ii in wordcount if len(ii) < 2]
    
    title = i.getTitle()
    title = re.findall(tickerRegex, title)
    
    if (len(countx) > 5 or len(county) > 5) and len(title) == 0:
        newsflag = True
        i.makeNewsWire()
    else:
        newsflag = False
    
    df2 = pd.DataFrame([{ "Number": i.getNumber(), "Ticker" : ticker, "Date": i.getDate(), "Title": i.getTitle(), "Wordcount": len(wordcount),
               "Ticker Count": len(z), "3Letter": len(countx), "4Letter": len(county), "Newswire": newsflag, "Object": i}])
    
    df = pd.concat([df, df2], ignore_index=True)

df = df.astype({'Date': 'datetime64[ns, UTC]'})
df['Month'] = df['Date'].dt.month
df['Year'] = df['Date'].dt.year
df['Day'] = df['Date'].dt.dayofyear
for index, row in df.iterrows():
    if row['Year'] == 2020:
        df.loc[index, 'Day'] = int(df.loc[index, 'Date'].dayofyear + 365)
    else:
        df.loc[index, 'Day'] = int(df.loc[index, 'Date'].dayofyear)
df = df.sort_values(by='Date')

print('complete')
```

```python
#nw[1] is newswires
#nw[0] is non-newswires

nw = df.groupby('Newswire')    
nw = [nw.get_group(x) for x in nw.groups]


# for index, row in nw[1].iterrows():
#     senty = re.sub(r"(?<=\d)\.(?=\d)", ",",row["Object"].getText())
#     senty = re.sub(r"(?<=[A-Z]{3})\.(?=AX)", ",", senty)
#     nwregex = r"[^.?!*]*(?<=[.?\s!])" + ticker + r"(?=[\s.?!])[.AX]*[^.?!)*]*[.?!*]"
#     senty = re.findall(nwregex, senty)
#     #print(row["Number"])
#     #print(senty)
#     row["Object"].makeNWText(senty)

print('Complete')
```

```python
nw[0] = nw[0].sort_values(by='Date')

ans = [pd.DataFrame(y) for x, y in nw[0].groupby('Year', as_index=False)]
totalmonths = []

if len(ans) >1:
    nineteen = [pd.DataFrame(y) for x, y in ans[0].groupby('Month', as_index=False)]
    twenty = [pd.DataFrame(y) for x, y in ans[1].groupby('Month', as_index=False)]
    for i in nineteen:
        totalmonths.append(i)
    for i in twenty:
        totalmonths.append(i)
else:
    ans = [pd.DataFrame(y) for x, y in ans[0].groupby('Month', as_index=False)]
    for i in ans:
        totalmonths.append(i)
print('complete')
```

```python
lemma = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

totalwords = []
permonth = {}
valx = []
valy = []

for frames in totalmonths:
    
    periodmonth = frames.iloc[0]["Date"].month_name()
    periodyear = frames.iloc[0]["Date"].year
    period = periodmonth + '_' + str(periodyear)
    
    print(period)
    
    valx.append(period)
    valy.append(len(frames))
    
    arts = []
    
    for index, row in frames.iterrows(): 
        article = row["Object"].getText()
        article = article.split("(c)")[0]
        article = re.sub(r'http\S+', '', article)
        article = re.sub(r'[\w\.-]+@[\w\.-]+(\.[\w]+)+', '', article)

        cleaned = simple_preprocess(article) #gensim
        cleaned = [item for item in cleaned if item not in stop_words]
        cleaned = [lemma.lemmatize(word=w,pos='v') for w in cleaned]
        cleaned = [item for item in cleaned if len(item) > 2]
        totalwords.extend(cleaned)
        arts.append(cleaned)
        permonth[str(period)] = arts

print('\ndone')
```

```python
yearlong = df['Day'].value_counts()

yearlong = pd.DataFrame(yearlong)

yearlong['index1'] = yearlong.index

axisxticks = []
axisxlabels = []

f = date.fromordinal(yearlong['index1'].min())
td = timedelta(weeks=4.345)
axisxticks.append(f.isoformat()[5:])
for i in range(14):
    f = f + td
    axisxticks.append(f.isoformat()[5:])

td = timedelta(days=1)
f = yearlong['index1'].min()
axisxlabels.append(f)

end = date.fromordinal(f).month

while end != 7:
    f += 1
    end = date.fromordinal(f).month
axisxlabels.append(f)

for i in range(13):
    f += 30.415
    axisxlabels.append(int(f))

print('done')
```

```python
paths = glob.glob('Tickerdata/'+blah+'.csv')
datez = []

for i in paths:
    newdf = pd.read_csv(i)
    for index, row in newdf.iterrows():
        datetime_object = parser.parse(row['Date']) 
        if datetime_object.year == 2020:
            twen = date.toordinal(parser.parse("2020-01-01"))
            ordi = date.toordinal(datetime_object)
            datez.append(ordi-twen+365)
        else:
            ordi = date.toordinal(datetime_object)
            nine = date.toordinal(parser.parse("2019-01-01"))
            datez.append(ordi-nine)


returnz = pd.DataFrame(newdf['CLOSE'])/newdf['CLOSE'].max()*yearlong['Day'].max()
dailyret = newdf['CLOSE'].diff()
#vol = newdf['VOLUME']
returnz['Dates'] = pd.DataFrame(datez)
returnz['diff'] = dailyret
#returnz['volume'] = vol
returnz = returnz.set_index('Dates', drop = True)
print('complete')
```

```python
#classifying whether a month is positive or negative return
flag = None
closes = []
keytimes = []


for index, row in newdf.iterrows():
    stockdates = parser.parse(row['Date'])
    if flag == None:
        flag = stockdates
        keytimes.append(row['OPEN'])
    elif stockdates.month != flag.month:
        flag = stockdates
        keytimes.append(closes.pop())
        keytimes.append(row['OPEN'])
    else:
        closes.append(row['CLOSE'])
keytimes.append(closes.pop())

sentiment = []

while keytimes:
    month = keytimes.pop(0) - keytimes.pop(0)
    #print(month)
    if month < 0:
        #sentiment.append(0-month)
        sentiment.append(round(0-month, 2))
    elif month > 0:
        sentiment.append(round(0-month,2))

print('complete')
```

```python
wordcounts = []
poplist = []

palette = itertools.cycle(sns.color_palette("Set2", 15))

for i in permonth:
    wordcount = 0
    monthwords = []
    
    for j in permonth[i]:
        wordcount += len(j)
        monthwords.extend(j)
        
    popwords = dict(Counter(monthwords).most_common(10))
    poplist.append(popwords)
    wordcounts.append(wordcount)    

fig, axs = plt.subplots(nrows = 5, ncols = 3, figsize=(16,20))
fig.suptitle('10 Most Popular Words by Month for '+ticker, weight = 'bold', fontsize = 22, y = 0.91)
fig.subplots_adjust(wspace=0.3, hspace = 0.3)

count = 0
for row in range(5):
    for col in range(3):
        vals = wordcountfunc(poplist)
        ax = axs[row, col]
        ax.set_yticks(range(10))
        ax.set_yticklabels(vals[1], fontsize = 12)
        c = next(palette)
        ax.barh(vals[1], vals[0], align='center', color = c)
        ax.invert_yaxis()
        ax.set_title(valx[count], weight = 'bold', fontsize = 'large')
        count += 1
plt.savefig(savepath+'Top10.png')
plt.show()  

print()
plt.figure(figsize=(16, 10))
plt.title(ticker + " Articles Per Month", weight = 'bold', fontsize = 22)
plt.ylabel("Number of Articles")
plt.bar(valx,valy, align='center')
plt.xticks(rotation=90)
plt.savefig(savepath+'ArticlesPerMonth.png')
plt.show()

print()
plt.figure(figsize=(16,10))
plt.title(ticker +" Words Per Month", weight = 'bold', fontsize = 22)
plt.ylabel("Wordcount")
plt.bar(valx, wordcounts, align='center')
plt.xticks(rotation=90)
plt.savefig(savepath+'WordsPerMonth.png')
plt.show()
```

```python
#permonth.keys()
w2v_monthly = []

for i in permonth:
    if len(permonth[i]) >1:
        print('Now running W2V on ', i)
        model = Word2Vec(permonth[i], size=2, window=3, min_count=3, workers=5, iter=5)
        w2v_monthly.append(model)
print('done')
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import os

try:
    os.mkdir('Clusters')
except FileExistsError:
    pass

count = 0
for i in w2v_monthly:
    word_vectors = i.wv.syn0
    n_words = word_vectors.shape[0]
    vec_size = word_vectors.shape[1]
    print("#words = {0}, vector size = {1}".format(n_words, vec_size))
    
    ### N CLUSTERS CODE - GENERATE NUMBER OF CLUSTERS BASED ON WORDS?
    # - 100 words to a cluster
    nclust = int(n_words/100)
    print('Clusters: '+str(nclust))
    if nclust == 0:
        nclust += 1
    
    kmeans = KMeans(n_clusters=nclust, random_state=0)
    idx = kmeans.fit_predict(word_vectors)

    word_centroid_list = list(zip(model.wv.index2word, idx))
    word_centroid_list_sort = sorted(word_centroid_list, key=lambda el: el[1], reverse=False)
    try:
        os.mkdir('Clusters/'+ticker)
    except FileExistsError:
        pass
    
    name = "Clusters/"+ticker+"/" + ticker + "_Month_" + str(count) + "_Clusters.csv"
    file_out = open(name, 'w')
    #NEED TO WRITE OUT IF THE MONTH WAS POSITIVE OR NEGATIVE OPEN VS CLOSE. 
    ret = sentiment.pop(0)
    if ret > 0:
        sent = 'POSITIVE'
    elif ret < 0:
        sent = 'NEGATIVE'
    else:
        sent = 'NEUTRAL'
    file_out.write('Word'+','+'Cluster'+','"Month_"+str(count)+','+str(ret)+','+sent+'\n')
    for word_centroid in word_centroid_list_sort:
        line = word_centroid[0] + ',' + str(word_centroid[1]) + '\n'
        file_out.write(line)
    file_out.close()
    count += 1
print('COMPLETE')
```
